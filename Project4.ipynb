```python
import json

f = open("project4data-1.txt")
input_sequences = []
for line in f:
    input_sequences.append(line[0:-1])

f.close()
f = open("our_matrix_arr.json")
our_matrix = json.load(f)
f.close()
f = open("blosum50.json")
blosum50_matrix = json.load(f)
f.close()
```

```python
import GenomeAligner

sequence_aligner_ours = GenomeAligner.GenomeAligner(our_matrix, -1)
sequence_aligner_blosum = GenomeAligner.GenomeAligner(blosum50_matrix, -2)
alignment_scores_blosum = [[250 for i in range(8)] for i in range(8)]
alignment_scores_ours = [[610 for i in range(8)] for i in range(8)]

print("Blosum Matrix")

for i in range(len(input_sequences)):
    for j in range(len(input_sequences)):
        if i != j:
            alignment_scores_blosum[i][j] = sequence_aligner_blosum.calc_alignscore(input_sequences[i], input_sequences[j])
            print(f"{i+1},{j+1} Score: {alignment_scores_blosum[i][j]}")
            sequence_aligner_blosum.print_best_alignment_of_last_pair()
            print(" ")

print("Scores")
for line in alignment_scores_blosum:
    print(line)
print(" ")
print("Min, Max")
for i in range(len(input_sequences)):
    print(f"{i}: {min(alignment_scores_blosum[i])},{max(alignment_scores_blosum[i])}")


print("\nOur Matrix")

for i in range(len(input_sequences)):
    for j in range(len(input_sequences)):
        if i != j:
            alignment_scores_ours[i][j] = sequence_aligner_ours.calc_alignscore(input_sequences[i], input_sequences[j])
            print(f"{i+1},{j+1} Score: {alignment_scores_ours[i][j]}")
            sequence_aligner_ours.print_best_alignment_of_last_pair()
            print(" ")

print("Scores")
for line in alignment_scores_ours:
    print(line)
print(" ")
print("Min, Max")
for i in range(len(input_sequences)):
    print(f"{i}: {min(alignment_scores_ours[i])},{max(alignment_scores_ours[i])}")
```
